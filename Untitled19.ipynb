```python
#ques 1
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler, MinMaxScaler
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
print("data types",df.dtypes)
num_data = df.select_dtypes(include=['int64','float64'])
cat_data = df.select_dtypes(include=['object'])
print("Numerical data")
print(num_data)
print("Categorical data")
print(cat_data)
#handling missing values
print("Missing values")
print(df.isnull().sum())
#categorical data mode
#numerical data median/mean
median = df['Age'].median()
print("Mean",median)
df['Age'] = df['Age'].fillna(median)
df = df.drop(columns=['Cabin'])
#df = df.dropna(subset=['Cabin])
mode1= df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode1)
print(df.isnull().sum())

#outlier detection fare
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1 - (1.5*IQR)
upper_bound = Q3 + (1.5*IQR)
outliers = df[(df['Fare']<lower_bound) | (df['Fare']>upper_bound)]
print("Outliers")
print(outliers)
df['Fare'] = np.where(df['Fare']<lower_bound,lower_bound,df['Fare'])
df['Fare'] = np.where(df['Fare']>upper_bound,upper_bound,df['Fare'])

#label encoding
le = LabelEncoder() #categorical data where only two values are present like true/false, m/f etc
df['Sex'] = le.fit_transform(df['Sex'])
df = pd.get_dummies(df,columns=['Embarked']) #categorical data where more than two values

#standard scaling where outliers are present

'''scaler = StandardScaler()
df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])'''

#minmax scaler when normalization of data 0 to 1
scaler = MinMaxScaler()
df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])
```
